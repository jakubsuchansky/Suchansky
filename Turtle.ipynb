```python
from ColabTurtle.Turtle import *
import math

initializeTurtle(initial_speed=13,initial_window_size=(1000, 700))

bgcolor('#0a0a2e')  # Tmav modr

# Spirala
penup()
goto(200, 250)
pendown()

for i in range(70):
    r = int(127 + 127 * math.sin(i * 0.15))
    g = int(127 + 127 * math.sin(i * 0.15 + 2))
    b = int(127 + 127 * math.sin(i * 0.15 + 4))
    color(r, g, b)
    width(i%3+1)
    forward(i*3)
    left(92)

# Zlata hvzda
penup()
goto(850, 150)
pendown()
color('gold')
width(2)

for _ in range(5):
    forward(80)
    right(144)

# Tyrkysova spirala
penup()
goto(500, 550)
pendown()

for i in range(40):
    color(0, max(200 - i*3, 0), max(200 - i*2, 0))
    width(1)
    forward(i * 4)
    right(91)

# Kvet
penup()
goto(180, 550)
pendown()

for i in range(18):
    color(255, min(100 + i*5, 255), min(150 + i*3, 255))
    width(2)
    forward(60)
    right(30)
    forward(60)
    right(150)

# Zelena kytka - nahore uprostred
penup()
goto(500, 180)
pendown()

for i in range(8):
    color(50, min(150 + i*12, 255), 80)
    width(2)
    for _ in range(36):
        forward(12)
        right(10)
    right(45)

# Trojuhelnikova spirala - vetsi a barevna
penup()
goto(820, 300)
pendown()

for i in range(45):
    r = int(127 + 127 * math.sin(i * 0.2))
    g = int(127 + 127 * math.sin(i * 0.2 + 2))
    b = int(127 + 127 * math.sin(i * 0.2 + 4))
    color(r, g, b)
    forward(i * 4)
    right(121)

# Oranzova hvezda - vpravo dole
penup()
goto(800, 520)
pendown()

for i in range(8):
    color(255, min(100 + i*15, 255), 50)
    forward(50)
    right(135)
    forward(50)
    left(90)

#ramacek
penup()
goto(15, 15)
pendown()
color('white')
width(4)
def ram(x,y):
    goto(x,y)
    forward(20)
    right(180)
    forward(40)
    right(180)
    forward(20)
    right(90)
plus(15,15)
plus(985,15)
plus(985,685)
plus(15,685)
plus(15,15)
penup()
goto(500, 20)
```
